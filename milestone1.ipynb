```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Handle Missing Values
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# Encode Categorical
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])

# Scaling
scaler = MinMaxScaler()
num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = scaler.fit_transform(df[num_cols])

# Save
df.to_excel("transformed_output.xlsx", index=False)

print("Data Transformation Completed Successfully!")
```

```python


import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report


file_path = r"C:\Users\shett\OneDrive\Desktop\Business_sales_EDA.csv"
df = pd.read_csv(file_path, sep=";")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)


num_cols = df.select_dtypes(include=np.number).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])


scaler = MinMaxScaler()
df[num_cols] = scaler.fit_transform(df[num_cols])

print("Normalization Completed")



# Example: Create Revenue Feature (if Sales & Price exist)
if "Sales Volume" in df.columns and "Price" in df.columns:
    df["Revenue"] = df["Sales Volume"] * df["Price"]

# Example: High Sales Flag
if "Sales Volume" in df.columns:
    df["High_Sales"] = np.where(df["Sales Volume"] > df["Sales Volume"].mean(), 1, 0)

print("Data Enrichment Completed")



if "Sales Volume" in df.columns:
    
    conditions = [
        df["Sales Volume"] > df["Sales Volume"].quantile(0.66),
        df["Sales Volume"] < df["Sales Volume"].quantile(0.33)
    ]
    
    choices = ["Positive", "Negative"]
    
    df["PNN_Class"] = np.select(conditions, choices, default="Neutral")

print("PNN Classification Added")




df["PNN_Class"] = df["PNN_Class"].map({
    "Negative": 0,
    "Neutral": 1,
    "Positive": 2
})


features = df.select_dtypes(include=np.number).drop(columns=["PNN_Class"], errors="ignore")
target = df["PNN_Class"]
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)


model = RandomForestClassifier()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)

print("\nClassification Report:\n")
print(classification_report(y_test, y_pred))

df.to_excel("final_enriched_classified_output.xlsx", index=False)

print("\n Full Pipeline Completed Successfully!")
```
